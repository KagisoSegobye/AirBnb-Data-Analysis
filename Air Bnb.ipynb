### Task 5 Data Analysis

```python
# Identifying outliers in price

df = data[data['price'] <1500]

sns.boxplot(data=data, x='price')
plt.show()
```

```python
# Price distribution
data['price']
```

```python
plt.figure(figsize=(6,5))
sns.histplot(data=df, x='availability_365', bins=100)
plt.title('availability_365')
plt.ylabel("Frequency")
plt.show()
```

```python
df.columns
```

```python
df.groupby(by='neighbourhood_group')['price'].mean()
```

```python
## average price per bed
df.groupby(by='neighbourhood_group')['price per bed'].mean()
```

**Feature Engineering** 

```python
df['price per bed'] = df['price']/df['beds']
```

## Bi Variable Analysis

```python
# One variable dependency in another variable

df.columns
```

```python
# price dependencies on neighbourhood
sns.barplot(data=df, x='neighbourhood_group', y='price' ,hue='room_type')
plt.savefig('Price_vs_neighbourhood.png')
plt.show()

## Here we can see that big cities like manhattan prefer a hotel room compared to Staten Island which has a very small population
```

```python
plt.figure(figsize=(8,5))
plt.title("Locality and Review Dependency")
sns.scatterplot(data=df, x='number_of_reviews', y='price', hue='neighbourhood_group')
plt.show()

## Here we can see that if the price is high, chances of having a review are low
```

```python
df.dtypes
```

```python
sns.pairplot(data=df , vars=['price', 'minimum_nights', 'number_of_reviews', 'availability_365'], hue='room_type')
plt.show()
```

```python
# Geographical Distribution of Airbnb Listing

sns.scatterplot(data=df ,x='longitude',y='latitude', hue='room_type')
plt.savefig('Geographical_Distrubution.png')
plt.show()
```

```python
# heatmap -correlation of one variable with others for numerical column

corr = df[['latitude', 'longitude', 'price','minimum_nights','reviews_per_month', 'availability_365','beds']].corr()

plt.figure(figsize=(8,7))
sns.heatmap(data=corr,annot=True)
plt.savefig('Variable_Correlation_heatmap.png')
plt.show()
```
